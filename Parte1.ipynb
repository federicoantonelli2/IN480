## Funzioni API

```python
from larlib import *
```

**_larSplit_**: suddivide un intervallo reale in $ n $ parti uguali. Restituisce la lista di $ n+1 $ vertici  $ 1-dimensionali $.

```python
type(larSplit(1)(2))
```

**_larGrid_**: genera la rappresentazione LAR di celle di complessi $ 0- $ o $ 1-dimensionali $.

```python
type(larGrid(2)(1))
```

**_larGridSkeleton_**: utile per assemblare le varie componenti dello scheletro di una griglia. In base alla dimensione passata in input restituisce lista di indici dei vertici, di spigoli o di facce.

```python
type(larGridSkeleton([1,1])(1))
```

**_larCuboids_**: utile per la costruzione di griglie (iper-)cuboidali. Il complesso generato Ã¨ un $ solido $, ed Ã¨ possibile includere le celle di tutte le dimensioni. Restituisce una tupla di vertici e facce.

```python
type(larCuboids([1,1,1]))
```

**_gridSkeletons_**: restituisce la lista di k-complessi cellulari con $ 0â‰¤k â‰¤ d $, dove $d$ Ã¨ la dimensione.

```python
type(gridSkeletons([1,1,1]))
```

**_larModelProduct_**: un modello LAR visualizzabile Ã¨ una coppia $(geometria,topologia)$, dove $geometria$ Ã¨ la lista di vertici e $topologia$ Ã¨ la lista delle celle in una qualche dimensione $d$. Questa funzione prende in input una coppia di modelli LAR e restituisce il modello del loro prodotto Cartesiano.

```python
mod_1= larSplit(1)(4), larGrid(4)(1)
mod_2= larSplit(3)(4), larGrid(4)(1)
#rettangolo=larModelProduct([mod_1,mod_2])
#rettangolo
#VIEW(SKEL_1(STRUCT(MKPOLS(rettangolo))))
type(larModelProduct([mod_1,mod_2]))
```

**_larSimplicialStack__**: utile per il calcolo dei simplessi. Creata una triangolazione, restituisce vertici, spigoli e triangoli.

```python
V,sim=larSimplexGrid1((2,2))
type(larSimplicialStack(sim))
```

**_larCuboidsFacets_**: estrae le facce da complessi cuboidali. Restituisce una tupla di vertici e facce.

```python
type(larCuboidsFacets(larCuboids([1,1,1])))
```
